```python
# -*- coding: utf-8 -*-

# La librairie numpy contient la majorit des lments mathmatiques utiliss dans ce travail.
import numpy as np
# La librairie matplotlib permet de gnrer des graphiques rapidement.
import matplotlib.pyplot as plt
# La librairie seaborn permet de faire des meilleurs prsentations graphiques.
import seaborn as sns

# Activation de l'utilisation de LaTeX pour le rendu du texte dans les graphiques
plt.rcParams.update({"text.usetex": True})
```

```python
# Compteur qui permet de gnrer les numros de figure automatiquement.
COMPTEUR_FIGURE = 1

# Dfinir une palette de couleurs pour les daltoniens (Flix-Antoine)
palette = sns.color_palette("colorblind")
sns.color_palette("colorblind") # Montrer la palette
```

## Exemple de graphique
Dans cette section, une fonction quelconque est d'abord crer, puis elle est affiche sous forme de graphique avec une police LaTeX.

```python
time = np.arange(-1.0, 1.0+0.01, 0.01)
velocity = np.cos(4*np.pi*time)
```

```python
fig, axe = plt.subplots(figsize=(16,8))
plt.plot(time, velocity, linestyle="solid", color=palette[2])
plt.title("Figure {} - ".format(COMPTEUR_FIGURE)+r"Graphic example in \TeX\ format $\rightarrow\displaystyle\sum_{n=1}^\infty\frac{-e^{i\pi}}{2^n}!$",
    fontsize=20, y=-0.1)
COMPTEUR_FIGURE += 1
axe.spines['left'].set_position('zero')
axe.spines['bottom'].set_position('zero')
axe.spines['right'].set_color('none')
axe.spines['top'].set_color('none')
axe.xaxis.set_ticks_position('bottom')
axe.yaxis.set_ticks_position('left')
axe.set_xlabel("Time [s]", fontsize=16, loc='right')
axe.set_ylabel("Velocity [m/s]", fontsize=16, loc='top')
axe.minorticks_on()
axe.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```
